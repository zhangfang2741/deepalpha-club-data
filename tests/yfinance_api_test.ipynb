```python
import yfinance as yf
import time
import random

# 1. 准备你想查询的板块
sectors = ['technology', 'financial-services', 'healthcare']

for sector_name in sectors:
    print(f"🚀 正在获取 {sector_name} 板块数据...")
    try:
        # 【核心修改】：不传任何 session，让 yfinance 默认用它最爱的 curl_cffi 处理
        tech_sector = yf.Sector(sector_name)
        
        for ind in tech_sector.industries:
            print(f"  - 行业: {ind['name']} -> Key: {ind['key']}")
            
    except Exception as e:
        print(f"❌ 获取 {sector_name} 失败，原因: {e}")
    
    # 2. 聪明的反爬策略：随机睡眠 3 到 5 秒
    # 固定时间的死循环容易被防火墙识别，让代码“装得更像人类在浏览器上点按”
    sleep_time = random.uniform(3.0, 5.0)
    print(f"😴 正在安全等待 {sleep_time:.2f} 秒，保护 IP...")
    time.sleep(sleep_time)
```
